# Data sources
Here I am going to practice how to 
- read csv data from the directory using sql
- setup the schema
- insert overwrite using sql

## Make sure that
The data file is in a Volume, so you can use spark sql freely

```python
%sql
-- plain select
SELECT * FROM csv.`dbfs:/Volumes/workspace/testing_schema/testing_volumne/testing_data/iris.csv`;
```

```python
%sql
-- read the data using a schema and header option
create or replace temporary view my_iris_date(
  sepal_len float,
  sepal_width float,
  petal_len float,
  petal_width float,
  species string
)
using csv
options (
path 'dbfs:/Volumes/workspace/testing_schema/testing_volumne/testing_data/iris.csv',
header 'true',
sep ','
);

select * from my_iris_date;

INSERT OVERWRITE DIRECTORY 'dbfs:/Volumes/workspace/testing_schema/testing_volumne/testing_data/iris2.csv'
USING CSV
OPTIONS ('header' = 'true')
SELECT * FROM my_iris_date;
```
